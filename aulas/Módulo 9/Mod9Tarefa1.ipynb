### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
# Data de aniversário
mes_dia_aniversario  = '05-31'

# Filtrar os dados para a data
sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])
sinasc_raw['mes_dia'] = sinasc_raw['DTNASC'].dt.strftime('%m-%d')
df_aniversario = sinasc_raw[sinasc_raw['mes_dia'] == mes_dia_aniversario]

# Calcular o peso médio dos bebês por sexo e faixa de escolaridade da mãe
df_aniversario.groupby(['SEXO', 'ESCMAE'])['PESO'].mean().to_frame()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
# Filtrar os dados para o ano de 2019
sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])
sinasc_2019 = sinasc_raw[sinasc_raw['DTNASC'].dt.year == 2019]

# Identificar o município onde nasceu o menor número de bebês em 2019
municipio_menos_bebes = sinasc_2019['munResNome'].value_counts().idxmin()
print(f'Município com menos nascimentos em 2019: {municipio_menos_bebes}')

# Filtrar os dados para o município identificado
sinasc_municipio = sinasc_2019[sinasc_2019['munResNome'] == municipio_menos_bebes]

# Calcular a idade média, máxima e mínima das mães nesse município
idade_mae_media = sinasc_municipio['IDADEMAE'].mean()
idade_mae_maxima = sinasc_municipio['IDADEMAE'].max()
idade_mae_minima = sinasc_municipio['IDADEMAE'].min()

print('Mães:')
print(f'Idade média das mães: {idade_mae_media}')
print(f'Idade máxima das mães: {idade_mae_maxima}')
print(f'Idade mínima das mães: {idade_mae_minima}')

# Calcular a idade média, máxima e mínima dos pais nesse município
# Supondo que a coluna com a idade dos pais seja 'IDADEPAI'
idade_pai_media = sinasc_municipio['IDADEPAI'].mean()
idade_pai_maxima = sinasc_municipio['IDADEPAI'].max()
idade_pai_minima = sinasc_municipio['IDADEPAI'].min()

# Espaçamento
print('-------------------------------------------')

print('Pais:')
print(f'Idade média dos pais: {idade_pai_media}')
print(f'Idade máxima dos pais: {idade_pai_maxima}')
print(f'Idade mínima dos pais: {idade_pai_minima}')
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc_marco = sinasc_raw[sinasc_raw['DTNASC'].dt.month == 3]

# Identificar o município onde nasceu o maior número de bebês em março
municipio_mais_bebes_marco = sinasc_marco['munResNome'].value_counts().idxmax()
print(f'Município com mais nascimentos em março: {municipio_mais_bebes_marco}')

# Filtrar os dados para o município identificado
sinasc_municipio_marco = sinasc_marco[sinasc_marco['munResNome'] == municipio_mais_bebes_marco]

# Calcular a quantidade média, máxima e mínima de filhos vivos nesse município
filhos_vivos_media = sinasc_municipio_marco['QTDFILVIVO'].mean()
filhos_vivos_maxima = sinasc_municipio_marco['QTDFILVIVO'].max()
filhos_vivos_minima = sinasc_municipio_marco['QTDFILVIVO'].min()

print(f'Quantidade média de filhos vivos: {filhos_vivos_media:.2f}')
print(f'Quantidade máxima de filhos vivos: {filhos_vivos_maxima}')
print(f'Quantidade mínima de filhos vivos: {filhos_vivos_minima}')

# Calcular a idade média, máxima e mínima dos pais nesse município
idade_pai_media = sinasc_municipio_marco['IDADEPAI'].mean()
idade_pai_maxima = sinasc_municipio_marco['IDADEPAI'].max()
idade_pai_minima = sinasc_municipio_marco['IDADEPAI'].min()

print(f'Idade média dos pais: {idade_pai_media:.2f}')
print(f'Idade máxima dos pais: {idade_pai_maxima}')
print(f'Idade mínima dos pais: {idade_pai_minima}')
```
