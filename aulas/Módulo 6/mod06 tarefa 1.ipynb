```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

df = pd.read_csv('db.csv')

df.head(10)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
amostragem = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']
df_barras = df[amostragem]

# Cria os subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Seta variáveis
titulos = ['APGAR1', 'APGAR5', 'Sexo', 'Gestação', 'Estado Civil da Mãe']

# Previne erro, transforma a matriz de eixos em uma lista plana para facilitar a iteração.
axes = axes.flatten()

# Criar um gráfico de barras para cada coluna
for ax, coluna, titulo in zip(axes, amostragem, titulos):
    df_barras[coluna].value_counts().plot(kind='bar', ax=ax)
    ax.set_title(titulo)
    ax.set_xlabel(coluna)
    ax.set_ylabel('Frequência')

# Remover o subplot vazio
fig.delaxes(axes[-1])

# Ajustar layout e exibir os gráficos
plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Agrupar os dados por 'CONSULTAS', calculando média de 'APGAR5'
apgar5_medio_por_consultas = df.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(apgar5_medio_por_consultas['CONSULTAS'], apgar5_medio_por_consultas['APGAR5'], color='skyblue')
plt.xlabel('Número de Consultas')
plt.ylabel('APGAR5 Médio')
plt.title('APGAR5 Médio por Número de Consultas')
plt.xticks(apgar5_medio_por_consultas['CONSULTAS'])  # Mostrar todos os valores de CONSULTAS no eixo x
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Exibir o gráfico
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
amostragem = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

# Seta variáveis
titulos = ['Sexo', 'Gestação', 'Estado Civil da Mãe', 'Consultas']

# Previne erro, transforma a matriz de eixos em uma lista plana para facilitar a iteração.
axes = axes.flatten()

# Criar um gráfico de barras para cada coluna
for ax, coluna, titulo in zip(axes, amostragem, titulos):
    df[coluna].value_counts().plot(kind='bar', ax=ax)
    ax.set_title(titulo)
    ax.set_xlabel("")
    ax.set_ylabel('Frequência')

# Ajustar layout e exibir os gráficos
plt.tight_layout()
plt.show()
```
